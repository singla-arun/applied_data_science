```python
# Query and merge series objects together
# Can be queried by either the index position or the index label
# if you dont given an index, the position and the label are effectively the same value
```

```python
import pandas as pd
student_classes = { "Alice": "Physics", "Bob": "Math", "Chris": "Chemistry"} 
```

```python
series = pd.Series(data = student_classes)
series
```

```python
series['Bob']
```

```python
series.loc['Bob']
```

```python
series.iloc[1]
```

```python
series[1]
```
