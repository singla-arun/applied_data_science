### Key Concepts
- **Strong connectivity:** Every node can reach every other node (directed paths)
- **Weak connectivity:** Underlying undirected graph is connected
- **Eccentricity:** Maximum shortest path distance from a node to any other node
- **Diameter:** Maximum eccentricity; **Radius:** Minimum eccentricity

Imports NetworkX for graph analysis. The email network is stored in a text file with three columns: sender, recipient, and Unix timestamp. ### Question 1

Using networkx, load up the directed multigraph from `assets/email_network.txt`. Make sure the node names are strings.

*This function should return a directed multigraph networkx graph.*

```python
import networkx as nx
def answer_one():
    
    # YOUR CODE HERE
        # YOUR CODE HERE
    path = "assets/email_network.txt"
    #G= nx.read_edgelist()

    G = nx.read_edgelist(path=path, data=[ ('time', float)], 
                            create_using=nx.MultiDiGraph())

    G.is_directed(), G.is_multigraph()
    #display (G.nodes())
    #display(list(G.edges(data = True)))
    #display(G.edges())
    return G
    raise NotImplementedError()
```

### Question 6

Using the NetworkX functions `strongly_connected_components` and `subgraph`, find the subgraph of nodes in the largest strongly connected component. 
Call this graph G_sc.

*This function should return a networkx MultiDiGraph named G_sc.*

```python
def answer_six():
    # YOUR CODE HERE
    G = answer_one()
    component_list = sorted(nx.strongly_connected_components(G))
    component_list

    sorted_list = sorted(component_list, key = len, reverse = True)
    longest = sorted_list[0]
    longest

    G_sc=nx.subgraph(G, nbunch = longest)
    #raise NotImplementedError()
    return G_sc
    raise NotImplementedError()
```

### Question 11

Which node in G_sc has the most shortest paths to other nodes whose distance equal the diameter of G_sc?


For the node with the most such shortest paths, how many of these paths are there?


*This function should return a tuple (name of node, number of paths).*

```python
# for all nodes A:
#     A_path_count = 0
#     for all nodes B not equal to A:
#        if the shortest path from A to B is equal the diameter:
#            A_path_count += number of shortest paths from A to B
#     keep (A, A_path_count) if is has the largest A_path_count
# return (A, A_path_count) for the node A with the largest A_path_count

def answer_eleven():
    # YOUR CODE HERE
    #from icecream import ic

    G_sc = answer_six()

    d_Gsc = nx.diameter(G_sc)
    display(d_Gsc)

    node_list = list(G_sc.nodes())
                    
    #display(node_list)

    node_list_out = []
    for node in node_list:
        path_cnt = 0 
        revised_node_list = [ target_node for target_node in node_list if target_node not in  [node]]

        #display(revised_node_list)
        for target_node in revised_node_list:
            if target_node ==node:
                pass
            else:
                shortest_path_length = nx.shortest_path_length(G_sc, source =node, target = target_node ) 
                #ic(shortest_path_length)
                
                if shortest_path_length == d_Gsc:
                    path_cnt = path_cnt+1
                    #display("pth_cnt_incremented")
        node_list_out.append((node, path_cnt))

        #ic(node_list_out)

    display(node_list_out)

    node_list_sorted = sorted(node_list_out, key = lambda x: x[1], reverse = True)
    a=node_list_sorted[0]
    return a    
    raise NotImplementedError()
```
