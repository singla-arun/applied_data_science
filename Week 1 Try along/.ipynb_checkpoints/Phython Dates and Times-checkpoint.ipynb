```python
import datetime as dt
import time as tm
```

```python
tm.time() # time in seconds/microseconds since Jan 1 19790
```

```python
# convert the time into datetime using the datetime object, datetime indicates date and time

dtnow = dt.datetime.fromtimestamp(tm.time())
dtnow
```

```python
dtnow.date()
```

```python
delta = dt.timedelta(days =100)
delta
```

```python
previousdatetime = dtnow - delta #find a datimetime 100 days before today
previousdatetime
```
