```python
target_transformed = pd.get_dummies(kyphosis_dataset['Kyphosis'])
target_transformed
target_transformed.index
kyphosis_dataset.index
#type(target_transformed)
type(kyphosis_dataset)

```

```python
#kyphosis_dataset.merge(target_transformed, how = 'inner' left_index =True,right_index = True )
merged_dataset = kyphosis_dataset.merge(right =target_transformed, how = 'inner', left_index = True, right_index =True)
```

```python
merged_dataset.drop(labels ='Kyphosis', axis =1, inplace =True)
```

```python
merged_dataset['Age'].aggregate(['min', 'max', 'mean'])
```

```python
merged_dataset['Age']=merged_dataset['Age'].astype(dtype =float)
```

```python
def months_to_years(months):
    years=months/12
    return years
```

```python
merged_dataset['Age in Years'] =merged_dataset['Age'].apply(months_to_years)
```
