```python
class Person:
    def __init__(self):
        self.name = 'Arun'
        self.location = 'Nashville'
        
    department = 'School of Information' #class variable which is shared across all instances of the classes
    
    def set_name(self, new_name):  #To have access to the instance upon which the method is being invoked you must pass in self as a parameter 
        self.name = new_name #istance variables of the object can be accessed by preceding them with self.
    def set_location(self, new_location):
        self.location = new_location
```

```python
new_person = Person()
```

```python
print(new_person.name )
print(new_person.location)
```
