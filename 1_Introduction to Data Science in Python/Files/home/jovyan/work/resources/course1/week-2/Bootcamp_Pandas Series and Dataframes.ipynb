```python
#Series are built on top of numpy arrays
import numpy as np
import pandas as pd

list_1 = ['a', 'b', 'c', 'd']
labels = [1,2,3,4]

ser_1 =pd.Series(data = list_1, index = labels)
arr_1 = np.array([1,2,3,4])
ser_2=pd.Series(arr_1)
display(ser_2)

dict_1 = { 1:'a', 2: 'b', 3:'c', 4: 'd'}
ser_3 = pd.Series(dict_1)          
display(ser_3)          
```

```python
ser_2 +ser_2
```
