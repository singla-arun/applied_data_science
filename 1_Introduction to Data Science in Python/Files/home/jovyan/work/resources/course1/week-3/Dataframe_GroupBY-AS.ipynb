```python
#Ratio Scale: Units are equally spaced. Height and weight. Mathematical operations are valid
#Interval scale: Units are equally spaced but there is no true zero.Operation such as multiplication and division are not valid. Example:  
#Temperatature as measured in Celcisus. Direction on compass
#Ordinal Scale: The order of units is important but not evenly spaced; Letter grades such as A+, A, A-
#Nominal scale: Names of teams 
```

```python
#Converting scales with Pandas
# first change to categorial then ordinal 

import pandas as pd

df = pd.DataFrame( data = ['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-'], index = ['excellant', 'excellant', 'excellant','good', 'good', 'good', 'ok', 'ok', 'ok'], columns = ['Grades'])
```

```python
df['Grades'].astype(dtype = 'category').head()
```

```python
my_categories =  pd.CategoricalDtype( categories = ['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-'],ordered = True)
```

```python
grades = df['Grades'].astype(dtype = my_categories)
```
