## Question 1

````python
def proportion_of_education():
    import pandas as pd
    mothers_education = pd.read_csv(filepath_or_buffer = 'assets/NISPUF17.csv' ,  index_col = None, header = 0, usecols = ['EDUC1', 'SEQNUMC'])

    #Find total number of records 
    mothers_education['EDUC1'].replace(to_replace = {1: "less than high school", 2:"high school", 3:"more than high school but not college", 4: "college"}, inplace = True)

    percentages = mothers_education.groupby(by = 'EDUC1', axis = 0).count()/mothers_education['SEQNUMC'].count()
    #print(percentages)
    #print(type(percentages.T))
    returnable = percentages['SEQNUMC'].to_dict()
    #returnable    

    return (returnable)
    raise NotImplementedError()
````


````python
def chickenpox_by_sex():
    # YOUR CODE HERE
    #HAD_CPOX --did the child ever have chicken pox, Yes (1), No(2), Others
    #P_NUMVRC - Total number of varicella doses, 0, 1, 2, NA
    # SEx - Sex of the child , Male (1), Female (2)

    df= pd.read_csv(filepath_or_buffer = 'assets/NISPUF17.csv' ,  index_col = None, header = 0, usecols =['SEX', 'HAD_CPOX', 'P_NUMVRC'] )
    df_male = df [df['SEX']==1]

    df_male_cleaned =df_male.dropna()
    mask =  (df_male_cleaned['P_NUMVRC']> 0.0) & (df_male_cleaned['HAD_CPOX'].isin([1,2]))
    df_male_vaccinated = df_male_cleaned[mask]

    import numpy as np

    df_male_vaccinated['infected'] =  list(np.where( df_male_vaccinated['HAD_CPOX'] ==1  ,'Infected at least once','Uninfected') )
    #print(df_male_vaccinated)
    df_male_vaccinated.drop(['SEX','P_NUMVRC'], axis = 1, inplace =True)
    df_male_grouped = df_male_vaccinated.groupby(by = ['infected']).count()
    df_male_grouped
    dict = {}
    dict['male'] = df_male_grouped.loc['Infected at least once']['HAD_CPOX']/df_male_grouped.loc['Uninfected']['HAD_CPOX']
    #dict 

    df_female = df [df['SEX']==2]

    #print(df_male.head(5))
    df_female_cleaned =df_female.dropna()
    mask =  (df_female_cleaned['P_NUMVRC']> 0.0) & (df_female_cleaned['HAD_CPOX'].isin([1,2]))
    df_female_vaccinated = df_female_cleaned[mask]

   # import numpy as np
    df_female_vaccinated['infected'] =  list(np.where( df_female_vaccinated['HAD_CPOX'] ==1  ,'Infected at least once','Uninfected') )
    #print(df_female_vaccinated)
    df_female_vaccinated.drop(['SEX','P_NUMVRC'], axis = 1, inplace =True)
    df_female_grouped = df_female_vaccinated.groupby(by = ['infected']).count()
    df_female_grouped
    dict['female'] = df_female_grouped.loc['Infected at least once']['HAD_CPOX']/df_female_grouped.loc['Uninfected']['HAD_CPOX']
    print(dict)
    return(dict)
    raise NotImplementedError()
````
