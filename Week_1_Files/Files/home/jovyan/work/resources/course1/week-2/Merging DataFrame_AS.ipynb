```python
# Inner Join
import pandas as pd
#Create two dataframes
df_staff = pd.DataFrame( [ { 'Name': 'Kelly', 'Role': "Director of HR"}, 
                                                                 { 'Name': 'Sally', 'Role': "Course Liason"},
                                                                 { 'Name': 'James', 'Role': "Grader"}
                                                                ])
df_staff = df_staff.set_index(keys ='Name')
print(df_staff)
```

```python
#Create the second
df_students = pd.DataFrame( [ { 'Name': 'James', 'School': "Business"}, 
                                                                 { 'Name': 'Mike', 'School': "Law"},
                                                                 { 'Name': 'Sally', 'School': "Engineering"}
                                                                ])
df_students = df_students.set_index(keys ='Name')
```

```python
# if we want the union 
print(pd.merge(left = df_staff, right = df_students, how = 'outer', left_on = 'Name', right_on = 'Name'))
print(pd.merge(left = df_staff, right = df_students, how = 'outer', left_index = True, right_index = True))
```

```python
print(pd.merge(left = df_staff, right = df_students, how = 'inner', left_index = True, right_index = True))
```

```python
#Examples of Set Addition
#1. List of all staff regardless of whether they are students or not. If they were students, we would want to get student details
#2 List of all students regardless of whehter they were staff or not
pd.merge(left = df_staff, right = df_students, how = 'left', left_index = True, right_index = True)
```

```python
pd.merge(left = df_staff, right = df_students, how = 'right', left_index = True, right_index = True)
```
