```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/cwurData.csv')
df.head()
```

```python
# define a function to categorize the rank into rank levels
def categorize_rank(rank): 
    if rank <=100: 
        return "First Tier"
    elif (rank >100) & (rank <=200):
        return "Second Tier"
    elif (rank >200) & (rank <=300):
        return "Third Tier"
    return "Other Top Tier"   
```

```python
df['rank_tier'] = df['world_rank'].apply( lambda x: categorize_rank(x))
```

```python
df.pivot_table(index = 'country',columns = 'rank_tier', values = 'score', aggfunc = [np.mean, np.max], margins =True).head()
```
