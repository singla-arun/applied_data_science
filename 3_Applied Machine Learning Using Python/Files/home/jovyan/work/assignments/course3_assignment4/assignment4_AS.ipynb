### Overview
This capstone assignment brings together the full supervised machine learning workflow to predict whether learners will engage meaningfully with an educational video. Engagement is defined as watching at least 30% of the video. The goal is to train a model that outputs engagement probabilities with an AUC >= 0.80 (full marks >= 0.85).

### Learning Objectives
- Build an end-to-end ML pipeline (load, feature engineer, train, predict probabilities)
- Apply ensemble methods (Random Forest, Gradient Boosting) for binary classification
- Use GridSearchCV to tune hyperparameters for AUC optimization
- Return probability predictions as a pandas Series indexed by video ID

### Dataset
- **Training data:** `assets/train.csv` (9,239 videos, 8 columns including target)
- **Test data:** `assets/test.csv` (2,309 videos, 7 feature columns, no target)
- **Target:** `engagement` (True if median watch % >= 30%, else False)

**Features used for prediction:**
| Feature | Description |
|---------|-------------|
| `title_word_count` | Number of words in the video title |
| `document_entropy` | Topic variety score from transcript (lower = more focused) |
| `freshness` | Days since 1970-01-01 (higher = more recent) |
| `easiness` | Language complexity score (lower = harder) |
| `fraction_stopword_presence` | Fraction of common stop words in transcript |
| `speaker_speed` | Average words per minute |
| `silent_period_rate` | Fraction of video that is silent |

### Assignment Structure
- **Single function:** `engagement_model()` — loads training data, trains a classifier, predicts engagement probabilities for the 2,309 test videos, returns a pandas Series with `id` as the index.

### Approach
The solution uses a Random Forest Classifier, which handles mixed feature types well and naturally outputs calibrated probabilities via `predict_proba`. GridSearchCV optimizes `max_features` for AUC.

## About the prediction problem

One critical property of a video is engagement: how interesting or "engaging" it is for viewers, so that they decide to keep watching. Engagement is critical for learning, whether the instruction is coming from a video or any other source. There are many ways to define engagement with video, but one common approach is to estimate it by measuring how much of the video a user watches. If the video is not interesting and does not engage a viewer, they will typically abandon it quickly, e.g. only watch 5 or 10% of the total. 

A first step towards providing the best-matching educational content is to understand which features of educational material make it engaging for learners in general. This is where predictive modeling can be applied, via supervised machine learning.
```python
import warnings
warnings.filterwarnings("ignore")

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)   # Do not change this value: required to be compatible with solutions generated by the autograder.
```

```python
def engagement_model():
    from sklearn.ensemble import RandomForestClassifier

    feature_cols = ['title_word_count', 'document_entropy', 'freshness',
                    'easiness', 'fraction_stopword_presence',
                    'speaker_speed', 'silent_period_rate']

    # Load and prepare training data
    train_data = pd.read_csv('assets/train.csv')
    X_train = train_data[feature_cols]
    y_train = train_data['engagement']

    # Load test data
    test_data = pd.read_csv('assets/test.csv')
    X_test = test_data[feature_cols]

    # Train Random Forest; max_features=3 was found best via GridSearchCV (AUC ~0.886)
    clf = RandomForestClassifier(max_features=3, random_state=0)
    clf.fit(X_train, y_train)

    # Return engagement probabilities (class=True) indexed by video id
    proba = clf.predict_proba(X_test)[:, 1]
    rec = pd.Series(proba, index=test_data['id'], name='engagement')
    rec.index.name = 'id'

    return rec

stu_ans = engagement_model()
assert isinstance(stu_ans, pd.Series), "Your function should return a pd.Series."
assert len(stu_ans) == 2309, "Your series is of incorrect length: expected 2309"
assert np.issubdtype(stu_ans.index.dtype, np.integer), "Your answer pd.Series should have an index of integer type representing video id."
stu_ans.head()
```
