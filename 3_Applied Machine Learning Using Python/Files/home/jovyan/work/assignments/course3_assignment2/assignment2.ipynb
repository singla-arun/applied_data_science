```python
# Exploratory cell: test polynomial degree 12 on train/test data (non-function version)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Lasso, LinearRegression

poly = PolynomialFeatures(degree=12)
X_train_poly = poly.fit_transform(X_train.reshape(-1,1))
X_test_poly = poly.transform(X_test.reshape(-1,1))

lr = LinearRegression()
lr.fit(X_train_poly, y_train)
lr_score = lr.score(X_test_poly, y_test)

lasso = Lasso(alpha=0.01, max_iter=10000, tol=0.1)
lasso.fit(X_train_poly, y_train)
lasso_score = lasso.score(X_test_poly, y_test)

print("LinearRegression R2:", lr_score, "  Lasso R2:", lasso_score)
```

```python
def answer_four():
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.linear_model import Lasso, LinearRegression
    from sklearn.metrics import r2_score

    poly = PolynomialFeatures(degree=12)
    X_train_poly = poly.fit_transform(X_train.reshape(-1, 1))
    X_test_poly = poly.transform(X_test.reshape(-1, 1))

    # Non-regularized linear regression (prone to overfitting at degree 12)
    lr = LinearRegression()
    lr.fit(X_train_poly, y_train)
    lr_score = lr.score(X_test_poly, y_test)

    # Lasso regression adds L1 penalty to shrink coefficients -> better generalization
    lasso = Lasso(alpha=0.01, max_iter=10000, tol=0.1)
    lasso.fit(X_train_poly, y_train)
    lasso_score = lasso.score(X_test_poly, y_test)

    return (lr_score, lasso_score)

answer_four()
```
